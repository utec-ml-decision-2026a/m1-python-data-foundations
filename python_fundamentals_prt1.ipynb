## 1. Variables y Tipos de Datos

### ¿Qué es una Variable?

Una **variable** es un contenedor que almacena un valor. Piensa en ella como una caja etiquetada donde guardamos información para usar después.

### Características:
- **Nombre descriptivo**: `edad`, `nombre`, `temperatura`
- **Asignación con `=`**: El valor va a la derecha, la variable a la izquierda
- **Sin declaración de tipo**: Python descubre el tipo automáticamente
- **Reasignable**: Puedes cambiar su valor en cualquier momento

```python
print("="*60)
print("VARIABLES Y TIPOS DE DATOS")
print("="*60 + "\n")

# 1. Enteros (int)
edad = 25
numero = -100
print(f"Entero: edad = {edad}")
print(f"Tipo: {type(edad)}\n")

# 2. Flotantes (float)
altura = 1.75
temperatura = -3.5
print(f"Flotante: altura = {altura}")
print(f"Tipo: {type(altura)}\n")

# 3. Cadenas de texto (str)
nombre = "Maria"
profesion = 'Científica de Datos'
print(f"String: nombre = {nombre}")
print(f"Tipo: {type(nombre)}\n")

# 4. Booleanos (bool)
es_estudiante = True
tiene_experiencia = False
print(f"Booleano: es_estudiante = {es_estudiante}")
print(f"Tipo: {type(es_estudiante)}\n")

# 5. None - valor nulo
resultado = None
print(f"None: resultado = {resultado}")
print(f"Tipo: {type(resultado)}\n")

# Reasignación de variables
print("Reasignación:")
x = 10
print(f"Primero: x = {x}, tipo = {type(x)}")
x = "ahora soy texto"
print(f"Después: x = {x}, tipo = {type(x)}")
```

## 2. Operadores y Expresiones

### Tipos de Operadores

#### 2.1 Aritméticos: Operaciones matemáticas
- `+` Suma
- `-` Resta
- `*` Multiplicación
- `/` División (resultado flotante)
- `//` División entera (solo parte entera)
- `%` Módulo (residuo de la división)
- `**` Potencia

##### Orden y Precedencia:

Python tiene reglas para dar prioridad a ciertas operaciones aritméticas.

El acronimo PEMDAS es el que se utiliza para definir la prioridad:
- P: Paréntesis ")"
- E: Exponenetes "**"
- M: Multiplicación "*"
- D: División "/"
- A: Adición "+"
- S: Substracción "-"

```python
print("="*60)
print("OPERADORES ARITMÉTICOS")
print("="*60 + "\n")

a = 10
b = 3

print(f"a = {a}, b = {b}\n")
print(f"Suma:          a + b = {a + b}")
print(f"Resta:         a - b = {a - b}")
print(f"Multiplicación: a * b = {a * b}")
print(f"División:      a / b = {a / b}")
print(f"División entera: a // b = {a // b}")
print(f"Módulo:        a % b = {a % b}") # devuelve el resto (residuo) de una división aritmética entre dos números
print(f"Potencia:      a ** b = {a ** b}\n")

# Orden de operaciones (PEMDAS)
resultado = 2 + 3 * 4
print(f"Orden de operaciones: 2 + 3 * 4 = {resultado}")
print(f"(Se multiplica primero: 2 + 12 = 14)\n")

# Operadores con strings
print("Concatenación de strings:")
nombre = "Juan"
apellido = "Pérez"
print(f'nombre + apellido = {nombre + " " + apellido}')
print(f'nombre * 3 = {nombre * 3}\n')
```

#### 2.2 Operadores de Comparación: Comparar valores (devuelven True/False)
- `==` Igual
- `!=` No igual
- `<` Menor que
- `>` Mayor que
- `<=` Menor o igual
- `>=` Mayor o igual

```python
print("="*60)
print("OPERADORES DE COMPARACIÓN")
print("="*60 + "\n")

x = 5
y = 10

print(f"x = {x}, y = {y}\n")
print(f"x == y:  {x == y}    (¿son iguales?)")
print(f"x != y:  {x != y}    (¿son diferentes?)")
print(f"x < y:   {x < y}     (¿es x menor?)")
print(f"x > y:   {x > y}     (¿es x mayor?)")
print(f"x <= y:  {x <= y}    (¿x menor o igual?)")
print(f"x >= y:  {x >= y}    (¿x mayor o igual?)\n")

# Comparar strings
print("Comparación de strings:")
print(f"'abc' < 'xyz': {'abc' < 'xyz'}  (orden alfabético)\n")
```

#### 2.3 Operadores Lógicos: Combinar condiciones (devuelven True/False)
- `and` Ambas condiciones deben ser True
- `or` Al menos una condición debe ser True
- `not` Invierte el valor booleano

```python
print("="*60)
print("OPERADORES LÓGICOS")
print("="*60 + "\n")

edad = 25
tiene_licencia = True

print(f"edad = {edad}, tiene_licencia = {tiene_licencia}\n")

# AND - Ambas deben ser True
print("AND (ambas deben ser True):")
print(f"(edad >= 18) and (tiene_licencia): {(edad >= 18) and (tiene_licencia)}")
print(f"(edad >= 18) and (not tiene_licencia): {(edad >= 18) and (not tiene_licencia)}\n")

# OR - Al menos una debe ser True
print("OR (al menos una debe ser True):")
print(f"(edad < 18) or (tiene_licencia): {(edad < 18) or (tiene_licencia)}\n")

# NOT - Invierte el valor
print("NOT (invierte el valor):")
print(f"not tiene_licencia: {not tiene_licencia}")
print(f"not True: {not True}")
print(f"not False: {not False}")
```
